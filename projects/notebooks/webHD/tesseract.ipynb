```python
from matplotlib import pyplot as plt
import pytesseract
import cv2
```

```python
print(pytesseract.image_to_string(cv2.imread("images/test.png"), lang='chi_sim+eng'))
```

```python
img = cv2.imread('images/test.png')

d = pytesseract.image_to_data(img, output_type=pytesseract.Output.DICT, lang='chi_sim+eng')
print(d["text"])
n_boxes = len(d['level'])
for i in range(n_boxes):
    (x, y, w, h) = (d['left'][i], d['top'][i], d['width'][i], d['height'][i])
    cv2.rectangle(img, (x, y), (x + w, y + h), (255, 0, 0), 2)

plt.imshow(img)
plt.show()
```
